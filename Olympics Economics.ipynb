```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('olympics-economics.csv')

# Display the first few rows of the dataframe
print(df.head())  # This prints the first 5 rows of the dataframe

# To get more information about the dataframe, you can use:
print(df.info())  # Provides details like column names, data types, and non-null counts

# To display the entire dataframe (if it's small):
print(df)
```

```python
# Step 2: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 3: Check column names
print("Column names in the dataset:", df.columns)

# Step 4: Calculate the total medals (adjust the column names as needed)
total_gold = df['gold'].sum()  # Replace 'Gold' with the correct column name if necessary
total_silver = df['silver'].sum()  # Replace 'Silver' if necessary
total_bronze = df['bronze'].sum()  # Replace 'Bronze' if necessary

# Step 5: Create a bar graph for the total medals
medal_counts = {'Gold': total_gold, 'Silver': total_silver, 'Bronze': total_bronze}
medals = list(medal_counts.keys())
counts = list(medal_counts.values())

plt.figure(figsize=(8, 6))
plt.bar(medals, counts, color=['gold', 'silver', '#cd7f32'])  # Use appropriate colors for medals
plt.title('Total Medals')
plt.xlabel('Medal Type')
plt.ylabel('Total Count')
plt.show()
```

```python
# Step 2: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 3: Check the column names to ensure we use the right ones
print("Column names in the dataset:", df.columns)

# Step 4: Adjust 'country', 'gold', 'silver', and 'bronze' based on your actual column names
# Replace 'country', 'gold', 'silver', and 'bronze' with the correct column names in your dataset

# Step 5: Find the countries with the highest number of medals
try:
    # Highest Gold
    country_highest_gold = df.loc[df['gold'].idxmax(), 'country']
    highest_gold = df['gold'].max()
    
    # Highest Silver
    country_highest_silver = df.loc[df['silver'].idxmax(), 'country']
    highest_silver = df['silver'].max()
    
    # Highest Bronze
    country_highest_bronze = df.loc[df['bronze'].idxmax(), 'country']
    highest_bronze = df['bronze'].max()
    
    # Step 6: Create pie charts
    plt.figure(figsize=(15, 5))
    
    # Pie chart for Gold Medals
    plt.subplot(1, 3, 1)
    plt.pie([highest_gold, df['gold'].sum() - highest_gold], 
            labels=[country_highest_gold, 'Others'], 
            autopct='%1.1f%%', 
            colors=['gold', 'skyblue'])
    plt.title('Gold Medals Distribution')
    
    # Pie chart for Silver Medals
    plt.subplot(1, 3, 2)
    plt.pie([highest_silver, df['silver'].sum() - highest_silver], 
            labels=[country_highest_silver, 'Others'], 
            autopct='%1.1f%%', 
            colors=['silver', 'skyblue'])
    plt.title('Silver Medals Distribution')
    
    # Pie chart for Bronze Medals
    plt.subplot(1, 3, 3)
    plt.pie([highest_bronze, df['bronze'].sum() - highest_bronze], 
            labels=[country_highest_bronze, 'Others'], 
            autopct='%1.1f%%', 
            colors=['brown', 'skyblue'])
    plt.title('Bronze Medals Distribution')
    
    plt.tight_layout()
    plt.show()

except KeyError as e:
    print(f"Column not found: {e}. Please check the column names and adjust them accordingly.")
```

```python
# Step 2: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Load the dataset (assuming it's already loaded in df)
# df = pd.read_csv('path_to_your_file.csv') # Uncomment and modify this line if needed

# Step 4: Calculate the total medal count for each country
df['total_medals'] = df['gold'] + df['silver'] + df['bronze']

# Step 5: Select only the columns needed for the correlation analysis
# Replace 'gdp' and 'population' with the actual column names of economic indicators
correlation_data = df[['gdp', 'population', 'gold', 'silver', 'bronze', 'total_medals']]

# Step 6: Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Step 7: Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Step 8: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Analysis between Economic Indicators and Medal Counts')
plt.show()
```
